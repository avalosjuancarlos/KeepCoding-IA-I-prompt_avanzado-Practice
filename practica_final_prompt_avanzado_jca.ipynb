# Ejercicio 2: Sistema de Resúmenes y Puntos Clave con LangChain o OpenAI

## **Descripción:**

En este ejercicio, deberás implementar un sistema que, dado un texto, genere un resumen y liste los puntos clave utilizando LangChain o OpenAI.
Este ejercicio tiene como objetivo practicar y asimilar la programación vista durante las clases.

## Instrucciones:

- Seleccionar la herramienta: Escoge si utilizarás LangChain o OpenAI para
implementar el sistema.

- Implementación del sistema:
  - Crea un script que tome un texto como entrada.
  - Utiliza la herramienta seleccionada para generar un resumen del texto.
  - Extrae y lista los puntos clave del texto.
  - Prueba y validación: Prueba el sistema con varios textos de diferente longitud y complejidad para asegurar su efectividad y precisión.

## Evaluación:

- La correcta implementación del sistema utilizando la herramienta escogida.

- La calidad y precisión de los resúmenes generados.

- La relevancia y claridad de los puntos clave listados.

- La capacidad del sistema para manejar textos de diferentes longitudes y niveles de complejidad.

````python

def generate_prompt_resume_text(text):
  body =  f'''
  Obtener un "resumen" y listar los "puntos clave" del siguiente texto:
  {text}
  '''

  output_size = '''
  Limitar la respuesta a menos 200 palabras y retorna un "resumen" y "puntos clave"
  '''

  output_format = ''' Brinda la respuesta en "formato markdown" '''


  zero_shot_chain_of_thought = ''' "Piensa paso a paso" '''

  prompt = f'''
  {body}
  {output_size}
  {output_format}
  {zero_shot_chain_of_thought}
  '''

  return prompt
````

````python
def get_resume(client, prompt):
  model = "gpt-3.5-turbo"
  messages = [
    {
      "role": "system",
      "content":
        ''' "Actua como un experto en realizar resúmenes de documentos" '''
    },
    {
      "role": "user",
      "content": prompt
    }
  ]

  response = client.chat.completions.create(
    model=model,
    messages=messages
  )
  return response.choices[0].message.content
````
